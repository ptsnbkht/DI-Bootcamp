Why Data Visualization Is Important in Data Analysis:
Data visualization is crucial in data analysis because it helps to:

Simplify Complex Data: Large datasets can be difficult to interpret in raw form (e.g., spreadsheets or databases). Visualizations transform complex numbers into easily understandable visuals.

Identify Patterns and Trends: Charts and graphs can reveal patterns, trends, and correlations that might not be obvious in text-based data.

Support Decision Making: Clear visuals help stakeholders and decision-makers quickly grasp insights and take informed actions.

Communicate Findings Effectively: Data visuals make it easier to present results to both technical and non-technical audiences.

Detect Outliers and Anomalies: Unusual data points become easier to spot in a visual format.

Purpose of a Line Graph in Data Visualization:
A line graph is used to display data points connected by straight lines, typically over a continuous variable such as time.

Main Purposes:
Show Trends Over Time: Line graphs are ideal for showing how values change over time (e.g., sales growth per month).

Compare Multiple Variables: Multiple lines on the same graph can compare trends between different groups or categories.

Highlight Progression or Decline: They make it easy to see increases, decreases, or steady behavior in data.

```python

import matplotlib.pyplot as plt

# Temperature data for each day of the week
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Days of the week (for x-axis labels)
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

# Create the line plot
plt.plot(days, temperatures, marker='o') # Added marker for data points

# Add labels and title
plt.xlabel("Day")
plt.ylabel("Temperature (Â°F)")
plt.title("Temperature Variation Over a Week")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Monthly sales data
sales = [5000, 5500, 6200, 7000, 7500]
months = ['January', 'February', 'March', 'April', 'May']

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(months, sales, color='skyblue')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales Data for Retail Store')

# Display the plot
plt.tight_layout()
plt.show()
```

```python


import pandas as pd
import matplotlib.pyplot as plt
# Assume df is a pandas DataFrame with columns 'Do you have Anxiety?' and 'Choose your gender'
# Example DataFrame creation (replace with your actual data loading)
data = {
    'Do you have Anxiety?': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes'],
    'Choose your gender': ['Female', 'Male', 'Male', 'Female', 'Non-binary', 'Female', 'Male', 'Female']
}
df = pd.DataFrame(data)

# Filter the DataFrame to include only rows where 'Do you have Anxiety?' is 'Yes'
anxiety_df = df[df['Do you have Anxiety?'] == 'Yes']

# Calculate the proportion of students with anxiety for each gender
# Using value_counts(normalize=True) gives the proportions
anxiety_proportions = anxiety_df['Choose your gender'].value_counts(normalize=True).reset_index()
anxiety_proportions.columns = ['Gender', 'Proportion of Anxiety']

# Create the bar plot using Seaborn's barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Proportion of Anxiety', data=anxiety_proportions, palette='viridis')

# Add labels and title
plt.xlabel("Gender")
plt.ylabel("Proportion of Students with Anxiety")
plt.title("Proportion of Students Experiencing Anxiety Across Different Genders")

# Display the plot
plt.tight_layout()
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
# Assume df is a pandas DataFrame with columns 'Age' and 'Do you have Panic Attacks?'
# Example DataFrame creation (replace with your actual data loading)
data = {
    'Age': [18, 20, 22, 19, 21, 23, 18, 20, 22, 19],
    'Do you have Panic Attacks?': ['No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)

# Convert panic attack responses to numeric values (Yes=1, No=0)
df['Panic_Attack_Numeric'] = df['Do you have Panic Attacks?'].apply(lambda x: 1 if x == 'Yes' else 0)

# Create the scatter plot using Seaborn's scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Panic_Attack_Numeric', data=df, alpha=0.7, s=100, hue='Do you have Panic Attacks?') # s for point size, alpha for transparency

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Panic Attacks (Yes=1, No=0)")
plt.title("Relationship between Age and Panic Attack Occurrence")

# Adjust y-axis ticks to show only 0 and 1
plt.yticks([0, 1], ['No', 'Yes'])

# Display the plot
plt.tight_layout()
plt.show()
```
